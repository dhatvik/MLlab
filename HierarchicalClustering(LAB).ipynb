```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df=df.drop(['Class'],axis=1)
df.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10, 7))
plt.title("Dendograms AGNES")
dend = sch.dendrogram(sch.linkage(df, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
cl=cluster.fit_predict(df)
```

```python
X = df.values
plt.figure(figsize=(10, 7))  
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[cl==2, 0], X[cl==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('CLUSTERS AGNES')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10, 7))
plt.title("Dendograms Diana")
dend = sch.dendrogram(sch.linkage(df, method='complete'),orientation='bottom')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')  
cl=cluster.fit_predict(df)
```

```python
X = df.values
plt.figure(figsize=(10, 7))  
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[cl==2, 0], X[cl==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('CLUSTERS DIANA')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
